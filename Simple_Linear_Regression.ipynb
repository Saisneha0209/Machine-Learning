```python
import pandas as pd
data = pd.read_csv("Salary_dataset (1).csv")
data.head()
```

```python
data.drop("Unnamed: 0", axis = 1 , inplace = True)
data.head()
```

```python
x = data[["YearsExperience"]]
x = data.drop("Salary", axis = 1)
type(x)
```

```python
x=data.iloc[:,:-1]
type(x)
```

```python
y =data["Salary"]
type(y)
```

```python
y.ndim
```

```python
import matplotlib.pyplot as mlt
mlt.scatter(x,y,marker ="+")
mlt.title("BEHAVIOUR OF DATA", fontdict = {"size":20, "color" : "red"})
mlt.xlabel("X LABEL",fontdict = {"size":20, "color" : "green"})
mlt.ylabel("Y LABEL",fontdict = {"size":20, "color" : "blue"})
mlt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.2)
x_train.shape  #it is 2-d cause an array
```

```python
y_train.shape  #it is 1-d cause a series
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#y = mx + c
model.fit(x_train,y_train)
```

```python
#predict is an function jisme yeh vala logic chal raha h y=mx + c!!!!
model.predict([[1.4]])
```

```python
model.coef_  #yeh ek angle hai!!!!
```

```python
model.intercept_
```

```python
y = 9347.28011299 * 1.4 + 24429.80511912951
y
```

```python
mlt.scatter(x_train,y_train,label = "Training Data")
mlt.scatter(x_test,y_test,label ="Testing Data")
mlt.plot(x_train,model.predict(x_train), label ="BEST FIT LINE")
mlt.legend()
mlt.show()
```

```python
#BIAS yani ki error!!!!
#Jo actual data se kitna different answer hai thats called this!!!
```

```python
y_pred = model.predict(x_test)
#BIAS:-
evaluation = pd.DataFrame({
    "actual value" : y_test,
    "predicted value" : y_pred,
    "bias/error" : y_pred - y_test

})
evaluation
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("R2_SCORE:",r2_score(y_test, y_pred))
print("MEAN_ABSOLUTE ERROR IS:",mean_absolute_error(y_test,y_pred))
print("MEAN_SQUARED ERROR IS:",mean_squared_error(y_test,y_pred))

```

```python
import seaborn as sns
sns.heatmap(evaluation,annot = True, fmt= "0.2f")
```

```python
import joblib
joblib.dump(model,"SLR.joblib")
```
