```python
import tensorflow as tf
w = tf.constant([[5,4,3],
                 [2,2,3],
                 [1,2,6],
                 [2,1,2]])
x = tf.constant([[3],
                 [2],
                 [1]])
b = tf.constant([[1],
                 [5],
                 [6],
                 [0]])

y = tf.matmul(w, x) + b

print(y)
#broadcasting laws in this and do it like 3-D!!!!!
```
