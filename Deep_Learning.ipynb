```python
import tensorflow as tf
w = tf.constant([[5,4,3],
                 [2,2,3],
                 [1,2,6],
                 [2,1,2]])
x = tf.constant([[3],
                 [2],
                 [1]])
b = tf.constant([[1],
                 [5],
                 [6],
                 [0]])

y = tf.matmul(w, x) + b

print(y)
#broadcasting laws in this and do it like 3-D!!!!!
```

```python
import tensorflow as tf

w = tf.constant([5, 3, 2])
x = tf.constant([3,
                 1,
                 5])
b = tf.constant([[2]])   # make bias 2D as well
w = tf.reshape(w, (1, 3))
x = tf.reshape(x, (3, 1))
y = tf.matmul(w, x) + b
#without matmal what you can do is simply doo by MATRICES method!!!!
print(y)
```
