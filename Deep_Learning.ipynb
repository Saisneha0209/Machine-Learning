#Y = W * X + B

```python
import tensorflow as tf
w = tf.constant([[5,4,3],
                 [2,2,3],
                 [1,2,6],
                 [2,1,2]])
x = tf.constant([[3],
                 [2],
                 [1]])
b = tf.constant([[1],
                 [5],
                 [6],
                 [0]])

y = tf.matmul(w, x) + b

print(y)
#broadcasting laws in this and do it like 3-D!!!!!
```

```python
import tensorflow as tf

w = tf.constant([5, 3, 2])
x = tf.constant([3,
                 1,
                 5])
b = tf.constant([[2]])   # make bias 2D as well
w = tf.reshape(w, (1, 3))
x = tf.reshape(x, (3, 1))
y = tf.matmul(w, x) + b
#without matmal what you can do is simply doo by MATRICES method!!!!
print(y)
```

#HERE are the functions like tf.reduce_mean,tf.reduce_max,tf.reduce_min!!!

```python
import tensorflow as tf
a = tf.constant([0.33,0.66,0.34])
print(tf.random.uniform([30],minval = 5 , maxval = 12))
print("------------------------------------------------")
print(f"The mean is: {tf.reduce_mean(a)}")
print("--------------------------------------------")
print(f"The maximum is: {tf.reduce_max(a)}")
print("--------------------------------------------")
print(f"The minimum is: {tf.reduce_min(a)}")
```

#To check for greator,equal, or greater_equal,less_equal or smaller!!

```python
import tensorflow as tf
a = tf.constant([1244,45,14])
x= tf.greater(a,3)  #Here this means in a we are saying a is greater to 3!!
print(x)
```

```python
import tensorflow as tf
a = tf.constant([1244,45,14])
x= tf.equal(a,3)  #Here this means in a we are saying a is equal to 3!!
print(x)
```

```python
import tensorflow as tf
a = tf.constant([1244,45,14])
x= tf.less(a,3)  #Here this means in a we are saying a is less to 3!!
print(x)
```
